```python
import pandas as pd # Para evitar escrever pandas e trocar pela escrita apenas de pd para facilitar
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import yfinance as yf 

from pandas_datareader import data as web # Evita a escrita do data e troca pelo web
from datetime import datetime
```

```python
timestampStr = datetime.now().strftime("%Y-%m-%d")
paper = 'petr4'
```

```python
data = yf.download(
tickers=paper + '.SA', 
period="1d",
interval="1m")
# Plot the close prices
data.Close.plot()
plt.show()
```


df = web.DataReader(f'' + paper + '.SA', data_source='yahoo', start='2021-06-01', end=timestampStr, interval='m')
